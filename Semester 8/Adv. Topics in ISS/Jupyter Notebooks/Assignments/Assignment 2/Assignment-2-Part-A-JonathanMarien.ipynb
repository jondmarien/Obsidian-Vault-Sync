```python
# --- SETUP --- #
import pandas as pd

# Load the dataset
df = pd.read_csv(r"nvd_cve_2022.csv")
df.set_index('id', inplace=True)

# --- PART A + B --- #
# --- A: What are the columns? --- #
# --- B: What are the data types of each column? (I also show a sample of values from each column) --- #

# Display all column names
print("Columns in the dataset:")
print(df.columns.tolist())

# Display the data types of each column
print("\nData types of each column:")
print(df.dtypes)

# Show a sample of values from each column
print("\nSample values:")
print("summary:", df['summary'].iloc[0])
print("base_severity:", df['base_severity'].iloc[0])
print("impact_score:", df['impact_score'].iloc[0])
```

```python
# --- PART C --- #
# --- C: Drop Unnamed: 0 column from the dataset --- #

# Option 1: Drop the column named 'Unnamed: 0' explicitly if you know its exact name
# This is useful when you are certain about the column name
# df = df.drop(columns=['Unnamed: 0'])

# Option 2: Drop all columns whose names start with 'Unnamed'
# This is a safer and more robust approach to handle cases where there may be multiple unnamed columns,
# or if the column naming may change (e.g., after multiple CSV exports)

# Creates a boolean series to identify columns starting with 
# 'Unnamed' using `^` (start of string), then `~` negates it to select columns not matching. 
# This drops all columns whose names start with 'Unnamed' (e.g., 'Unnamed: 0') from the DataFrame.
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Display the updated dataset to confirm the column(s) have been removed
df.head()
```

```python
# --- PART D --- #
# --- D: Find the number of columns and rows (dimensionality) --- #
# Get the shape of the DataFrame
print("\nNumber of rows and columns (rows, columns):")
print(df.shape)
```

```python
# --- PART E --- #
# --- E: What does the column "references" show? --- # 

# --- Originally, I was confused, and I was looking for a column named "references" --- #
#print("\nSample 'references' value:")
#print(df['references'].iloc[0])

# --- I then realized that there must be a way to find a column with a similar name --- #
# --- So I used a for loop to find the column with a similar name --- #

# --- 1. Display all column names to find the references column --- #
print("Column names in the dataset:")
print(df.columns.tolist())

# --- 2. Find the correct references column and display a sample value --- #
# --- Try to find a column related to references (case-insensitive) --- #
references_col = None
for col in df.columns:
    if 'ref' in col.lower():
        references_col = col
        break

if references_col:
    print(f"\nSample value from the '{references_col}' column:")
    print(df[references_col].iloc[0])
else:
    print("\nNo column related to references found.")

# --- 3. Explanation (as a comment): --- #
# The 'references' column provides external URLs or resources for each CVE.
# These links point to advisories, patches, technical analyses, or other databases,
# which helps users find more information and mitigation steps for each vulnerability.

# --- Now that we know what the column is named, we can improve and simplify the code --- #
# --- We can use the column name to access the data more easily --- #
print("\nSample 'references' value:")
print(df['reference_data'].iloc[0])

# --- OR, we can print a certain number of reference values --- #
print("\nFirst 3 sample 'references' values:")
print(df['reference_data'].iloc[:3]) 
```

## <p style="color:purple"><b>Exercise 4:</b></p>
<p style="color:purple;"> Group cves based on base_severity and <br>
1- extract number of rows in each group. <br>
2- extract standard deviation value of base_score in each group.</p>

```python
# 1. Group by 'base_severity'
grouped = df.groupby('base_severity')

# 2. Number of rows (CVEs) in each group
count_per_group = grouped.size()
print("Number of rows in each `base_severity` group:")
print(count_per_group)

# 3. Standard deviation of `base_score` in each group
std_base_score_per_group = grouped['base_score'].std()
print("\nStandard deviation of `base_score` in each `base_severity` group:")
print(std_base_score_per_group)
```

## <p style="color:purple"><b>Exercise 5:</b></p>

<p style="color:purple;"> Find the minimun and maximum value of exploitability_score. <br>
Create a new column and name it "Exploitable". <br>
If base_score is above midrange, "Exploitable" is 1, otherwise it is 0. <br>
Use apply and lambda. <br></p>

<p style="color:purple;">Hint: The quantity obtained by adding the largest and smallest values and dividing by 2, statisticians call the midrange. </p>

```python
#a=df['exploitability_score'].max()
#b=df['exploitability_score'].min()
#(a+b)/2

# 1. Find min and max of exploitability_score
min_exploit = df['exploitability_score'].min()
max_exploit = df['exploitability_score'].max()

print(f"Minimum exploitability_score: {min_exploit}")
print(f"Maximum exploitability_score: {max_exploit}")

# 2. Calculate the midrange
midrange = (min_exploit + max_exploit) / 2
print(f"Midrange of exploitability_score: {midrange}")

# 3. Create the 'Exploitable' column using apply and lambda
df['Exploitable'] = df['base_score'].apply(lambda x: 1 if x > midrange else 0)

# Show a handful of rows to verify
print(df[['base_score', 'Exploitable']].iloc[:10])
```

## <p style="color:purple"><b>Exercise 6:</b></p>

<p style="color:purple;"> Extract the number of rows modified between May 5, 2022 to June 5, 2022. <br>
    Extract number of distinct CVEs reported between March 5, 2022 to April 5, 2022.</p>

```python
# --- 1. Convert date columns to datetime --- #
df['last_modified_date'] = pd.to_datetime(df['last_modified_date'])
df['publish_date'] = pd.to_datetime(df['publish_date'])
# --- 1a. Ensuring date columns are datetime with UTC timezone --- #
df['last_modified_date'] = pd.to_datetime(df['last_modified_date'], utc=True)
df['publish_date'] = pd.to_datetime(df['publish_date'], utc=True)

# --- 2. Create timezone-aware Timestamps for comparison --- #
# mod = last_modified_date
start_mod = pd.Timestamp('2022-05-05', tz='UTC')
end_mod = pd.Timestamp('2022-06-05', tz='UTC')
# --- 2a. Rows modified between May 5, 2022 and June 5, 2022 --- #
modified_rows = df[(df['last_modified_date'] >= start_mod) & (df['last_modified_date'] <= end_mod)]
print("Number of rows modified between May 5, 2022 and June 5, 2022:")
print(len(modified_rows))

# --- 3. Create timezone-aware Timestamps for comparison --- #
# pub = publish_date
start_pub = pd.Timestamp('2022-03-05', tz='UTC')
end_pub = pd.Timestamp('2022-04-05', tz='UTC')
# --- 3a. Distinct CVEs reported between March 5, 2022 and April 5, 2022 --- #
reported_cves = df[(df['publish_date'] >= start_pub) & (df['publish_date'] <= end_pub)]
num_distinct_cves = reported_cves['id'].nunique()
print("\nNumber of distinct CVEs reported between March 5, 2022 and April 5, 2022:")
print(num_distinct_cves)
```

## <p style="color:purple"><b>Exercise 7:</b></p>
<p style="color:purple;"> Draw a pie chart showing the distribution of base severity (Critical, High, Low, Medium). <br>
Hint: first group by base_severity, and find number of rows in each group, and use plot. </p>

```python
import matplotlib.pyplot as plt

# --- 1. Group by 'base_severity' and count the number of rows in each group --- #
severity_counts = df['base_severity'].value_counts()

print("Value counts for base_severity:")
print(severity_counts)

# --- 2. Draw the pie chart --- #
plt.figure(figsize=(6, 6))
severity_counts.plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    counterclock=False,
    colors=['yellow', 'orange', 'red', 'green']
)
plt.title('Distribution of Base Severity')
plt.ylabel('')  # Hide y-label for aesthetics
plt.show()
```
